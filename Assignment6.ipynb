```python
import threading
import numpy as np

arr = np.ones(100000000)

def work():
    global arr
    arr *= 2

threads = []
for i in range(4):
    t = threading.Thread(target=work)
    threads.append(t)
    t.start()

for t in threads:
    t.join()

print("Execution Completed")
```
